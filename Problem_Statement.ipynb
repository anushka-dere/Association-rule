```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
book = pd.read_csv("/content/book.csv")
book
```

```python
book.shape
```

Aprori algorithm

```python
frequent_itemsets_ap = apriori(book,min_support = 0.1)
```

```python
frequent_itemsets_ap
```

```python
len(frequent_itemsets_ap)
```

```python
frequent_itemsets_ap = apriori(book, min_support=0.1, use_colnames= True)
frequent_itemsets_ap
```

```python
frequent_itemsets_ap.sort_values("support", ascending = False)
```

```python
rules_ap = association_rules(frequent_itemsets_ap, metric = 'confidence', min_threshold = 0.4)
rules_ap
```

```python
rules_ap[(rules_ap.support>0.015) & (rules_ap.confidence > 0.4)].sort_values("confidence",ascending = False)
```

```python
rules_ap[(rules_ap.support>0.015) & (rules_ap.confidence > 0.4)].sort_values("confidence",ascending = False).shape
```

```python
rules_ap['lhs items'] = rules_ap['antecedents'].apply(lambda x:len(x))
rules_ap[rules_ap['lhs items']>1].sort_values('lift',ascending = False)
```

```python
rules_ap['antecedents_'] = rules_ap['antecedents'].apply(lambda a:','.join(list(a)))
rules_ap['antecedents_']
```

```python
rules_ap['consequents_'] = rules_ap['consequents'].apply(lambda a:','.join(list(a)))
rules_ap['consequents_']
```

```python
pivot =rules_ap[rules_ap['lhs items']>1].pivot(index = 'antecedents_',
                                               columns = 'consequents_', values = 'confidence')
```

```python
pivot
```

```python
sns.heatmap(pivot, annot = True)
plt.title('heat map - for confidence metric')
plt.yticks(rotation=0)
plt.xticks(rotation=90)
```
